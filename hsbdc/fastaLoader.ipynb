```python
from Bio import SeqIO
import pandas as pd

fasta_file = 'final_model_dataset/train.fa'

data = []
i=0
for record in SeqIO.parse(fasta_file, "fasta"):
    data.append({
        "ID": record.id,
        "Sequence": str(record.seq)
    })
    i += 1
    if i > 10000:
        break

df = pd.DataFrame(data)

df.to_csv('final_model_dataset/smallTrain.csv', index=False)
print(df)
```
